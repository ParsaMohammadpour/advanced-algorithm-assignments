### Reading file

```python
val_freq_lst = []
with open('C:\\Users\\parsa\\OneDrive\\Desktop\\Advanced-Algorithm\\Assignments\\1st-code-assignment\\scientificProject\\data\\huffman.txt', 'r') as file:
        num_symbols = int(file.readline().strip())
        for i in range(num_symbols):
            freq = int(file.readline().strip())
            val = str(i+1)
            val_freq_lst.append((val, freq))
```

### File Huffman 

```python
start_time = time.time_ns()
root = HuffmanAlgorithm.generate_tree(val_freq_lst)
end_time = time.time_ns()
print(f'execution time: {end_time - start_time} ns')
```

### Max & Min code length

```python
max_len = HuffmanAlgorithm.get_max_code_len(root)
min_len = HuffmanAlgorithm.get_min_code_len(root)
print(f'max code len: {max_len}')
print(f'min code len: {min_len}')
print(f'diff: {max_len - min_len}')
```
